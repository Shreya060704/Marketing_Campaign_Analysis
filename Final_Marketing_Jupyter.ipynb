```python
                                                              DATA CLEANING
```

```python
print(df.isnull().sum())

#CHECKS IF THERE IS A NULL DATA IN THE DATASET
```

```python
Null_row=df[df['Duration'].isnull()]
print(Null_row)
```

```python
df=df.dropna(subset=['Duration'])
```

```python
print(df.isnull().sum())
```

```python
print(df.dtypes)

#THIS IS DONE TO CHECK THE DATATYPE OF THE COLUMNS
```

```python
#CHANGING THE DATATPYE OF THE REQUIRED COLUMNS

df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y', errors='coerce')

df['Company'] = df['Company'].astype('category')


df['Campaign_Type'] = df['Campaign_Type'].astype('category')



df['Target_Audience'] = df['Target_Audience'].astype('category')


df['Channel_Used'] = df['Channel_Used'].astype('category')


df['Language'] = df['Language'].astype('category')


df['Customer_Segment'] = df['Customer_Segment'].astype('category')
```

```python
print(df.dtypes)

#PRINTS THE DATATYPES OF THE COLUMNS INCLUDING THE CHANGED ONES
```

```python
                                                         EXPLORATORY DATA ANALYSIS
```

```python
#COUNTS

campaign_type_distribution = df['Campaign_Type'].value_counts()
channel_type_distribution = df['Channel_Used'].value_counts()
Target_audience_distribution = df['Target_Audience'].value_counts()
Language_distribution = df['Language'].value_counts()
Customer_Segment_distribution = df['Customer_Segment'].value_counts()

print(campaign_type_distribution)
print("\n",channel_type_distribution)
print("\n",Target_audience_distribution)
print("\n",Language_distribution)
print("\n",Customer_Segment_distribution)
```

```python
                                                        CAMPAIGN TYPE ANALYSIS
```

```python
#DISTRIBUTION OF CAMPAIGN TYPES

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Campaign_Type', hue='Campaign_Type', palette='pastel', legend=False)
plt.title("Distribution of Campaign_Type")
plt.xlabel("Campaign Type")
plt.ylabel("Count")
plt.show()
```

```python
#CAMPAIGN TYPE GROUPED BY CONVERSION RATE

average_conversion_rate_by_campaign = df.groupby('Campaign_Type')['Conversion_Rate'].mean().sort_values(ascending=False)

print("Average Conversion Rate for each Campaign_Type:")
print(average_conversion_rate_by_campaign)

average_conversion_rate_by_campaign.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title('Average Conversion Rate by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Average Conversion Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#CAMPAIGN TYPE GROUPED BY ROI

average_roi_by_campaign = df.groupby('Campaign_Type')['ROI'].mean().sort_values(ascending=False)

print("Average ROI for each Campaign_Type:")
print(average_roi_by_campaign)

average_roi_by_campaign.plot(kind='bar', color='pink', figsize=(8, 5))
plt.title('Average ROI by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.show()
```

```python
#CAMPAIGN TYPE GROUPED ACQUISITION COST

average_acquisition_cost_by_campaign = df.groupby('Campaign_Type')['Acquisition_Cost'].mean().sort_values(ascending=False)

print("Average Acquisition Cost for each Campaign_Type:")
print(average_acquisition_cost_by_campaign)

average_acquisition_cost_by_campaign.plot(kind='bar', color='pink', figsize=(8, 5))
plt.title('Average Acquisition Cost by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Average Acquisition Cost')
plt.xticks(rotation=45)
plt.show()
```

```python
                                                                SEASONAL TRENDS ANALYSIS 
```

```python
df['Date'] = pd.to_datetime(df['Date'])


df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
monthly_trends = df.groupby('Month')[['Conversion_Rate', 'ROI','Engagement_Score']].mean().reset_index()


print(monthly_trends)
```

```python
#MONTHLY TRENDS IN CONVERSION RATE

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_trends, x='Month', y='Conversion_Rate', marker='o', label='Conversion Rate')
plt.title('Seasonal Trends in Conversion Rate (Monthly)')
plt.xlabel('Month')
plt.ylabel('Conversion Rate')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid()
plt.show()
```

```python
#SEASONAL TRENDS IN ROI

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_trends, x='Month', y='ROI', marker='o', label='ROI')
plt.title('Seasonal Trends in ROI (Monthly)')
plt.xlabel('Month')
plt.ylabel('ROI')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid()
plt.show()
```

```python
#SEASONAL TRENS IN THE ENGAGEMENT SCORE (MONTHLY)

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_trends, x='Month', y='Engagement_Score', marker='o', label='Engagement_Score')
plt.title('Seasonal Trends in Engagement_Score (Monthly)')
plt.xlabel('Month')
plt.ylabel('Engagement_Score')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid()
plt.show()
```

```python
Q1. Identify which Channel_Used delivers the highest ROI and Conversion_Rate.
```

```python
#CHANNEL WITH HIGHEST ROI

average_Channel_Used_by_campaign = df.groupby('Channel_Used')['ROI'].mean().sort_values(ascending=False)

print("Average Channel_Used for each ROI:")
print(average_Channel_Used_by_campaign)

#FACEBOOK HAS THE HIGHEST ROI
```

```python
#CHANNEL WITH HIGHEST CONVERSION RATE

average_Channel_Used_by_convertion = df.groupby('Channel_Used')['Conversion_Rate'].mean().sort_values(ascending=False)

print("Average Channel_Used for each Conversion_rate")
print(average_Channel_Used_by_convertion)

# GOOGLE ADS IS THE MOST USED CHANNEL
```

```python
Q2. Understand which Target_Audience segments respond best to campaigns.
```

```python
#TARGET AUDIENCE THAT RESPONDS BEST TO THE CAMPAIGNS

avg_target_Campaign=df.groupby('Target_Audience')['Campaign_Type'].count().sort_values(ascending=False)
print(avg_target_Campaign)


#MEN 25-34 RESPONSD BEST TO CAMPAIGNS
```

```python
Q3. Which Locations have the highest engagement and conversion?
```

```python
#LOCATION WITH HIGHEST ENGAGEMENT SCORE

engagement_rate_by_location = df.groupby('Location')['Engagement_Score'].mean()
print("\nAverage Engagement_Score by Location:")
print(engagement_rate_by_location)

#Chicago has the highest engagement_score
```

```python
#LOCATION WITH HIGHEST CONVERSION RATE

language_group = df.groupby('Language').agg({
    'Conversion_Rate': ['mean', 'std'], 
    'Acquisition_Cost': ['mean', 'std'],  
    'ROI': ['mean', 'std'],  
    'Engagement_Score': ['mean', 'std'],  
}).reset_index()

# Display the results
print(language_group)
```
